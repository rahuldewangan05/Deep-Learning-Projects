```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('./1661-0.txt', 'r', encoding='latin-1') as file:
    content = file.read()
```

```python
content=content.split("\n")
```

```python
content
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(content)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequence=[]
for i in content:
    tokenised=tokenizer.texts_to_sequences([i])[0]
    if(len(tokenised)==0):
        continue
    else:
        for j in range(1,len(tokenised)):
            input_sequence.append(tokenised[:j+1])
```

```python
input_sequence
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
padded_seq=pad_sequences(input_sequence,padding="pre")
```

```python
padded_seq
```

```python
X=padded_seq[:,:-1]
Y=padded_seq[:,-1]
```

```python
X
```

```python
X.shape
```

```python
Y
```

```python
Y.shape
```

```python
Y=pd.get_dummies(Y,dtype="int")
```

```python
Y.shape
```

```python
X.shape
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Embedding,LSTM
```

```python
model=Sequential()
model.add(Embedding(input_dim=8932,output_dim=100,input_length=19))
model.add(LSTM(150,return_sequences=True))   ### 150 is no. of nodes
model.add(LSTM(150))
model.add(Dense(8192,activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy",metrics=["accuracy"],optimizer="Adam")
```

```python
history=model.fit(X,Y,epochs=10,validation_split=0.2,verbose=1)
```

```python
text="this"
tokenised=tokenizer.texts_to_sequences([text])[0]
```

```python
padded=pad_sequences([tokenised],maxlen=19,padding="pre")
```

```python
pos=np.argmax(model.predict(padded))
```

```python
for word,index in tokenizer.word_index.items():
    if(index==pos):
        print(word)
```
